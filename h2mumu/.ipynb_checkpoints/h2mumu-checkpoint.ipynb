```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read in a csv to play with

```python
df = pd.read_csv('data/categorization_training.csv')
```

Make some cuts

```python
bkgdf = df[df.is_signal <= 0]
sigdf = df[df.is_signal > 0]
bkgdf.describe()
```

Make some plots

```python
%matplotlib inline
plt.xlim(0,200)
sns.distplot(sigdf.MET, kde=False, norm_hist=True, bins=100, color='green', hist_kws={'weights': sigdf.weight, 'alpha': 0.7})
sns.distplot(bkgdf.MET, kde=False, norm_hist=True, bins=100, color='blue', hist_kws={'weights': bkgdf.weight, 'alpha': 0.7})
```

```python
plt.xlim(0,125)
sns.distplot(sigdf.mu0_pt, kde=False, norm_hist=True, bins=100, color='green', hist_kws={'weights': sigdf.weight, 'alpha': 0.7})
sns.distplot(bkgdf.mu0_pt, kde=False, norm_hist=True, bins=100, color='blue', hist_kws={'weights': bkgdf.weight, 'alpha': 0.7})
```

```python
plt.xlim(0,100)
sns.distplot(sigdf.mu1_pt, kde=False, norm_hist=True, bins=100, color='green', hist_kws={'weights': sigdf.weight, 'alpha': 0.7})
sns.distplot(bkgdf.mu1_pt, kde=False, norm_hist=True, bins=100, color='blue', hist_kws={'weights': bkgdf.weight, 'alpha': 0.7})
```

```python
plt.xlim(0,125)
sigdfj = sigdf[sigdf.N_valid_jets > 0]
bkgdfj = bkgdf[bkgdf.N_valid_jets > 0]
sns.distplot(sigdfj.jet0_pt, kde=False, norm_hist=True, bins=500, color='green', hist_kws={'weights': sigdfj.weight, 'alpha': 0.7})
sns.distplot(bkgdfj.jet0_pt, kde=False, norm_hist=True, bins=500, color='blue', hist_kws={'weights': bkgdfj.weight, 'alpha': 0.7})
```

```python
plt.xlim(0,125)
sigdfj = sigdf[sigdf.N_valid_jets > 1]
bkgdfj = bkgdf[bkgdf.N_valid_jets > 1]
sns.distplot(sigdfj.jet1_pt, kde=False, norm_hist=True, bins=500, color='green', hist_kws={'weights': sigdfj.weight, 'alpha': 0.7})
sns.distplot(bkgdfj.jet1_pt, kde=False, norm_hist=True, bins=500, color='blue', hist_kws={'weights': bkgdfj.weight, 'alpha': 0.7})
```

```python
plt.xlim(50,250)
sigdfj = sigdf[sigdf.N_valid_jets > 1]
bkgdfj = bkgdf[bkgdf.N_valid_jets > 1]
sns.distplot(sigdfj.m_jj, kde=False, norm_hist=True, bins=1000, color='green', hist_kws={'weights': sigdfj.weight, 'alpha': 0.7})
sns.distplot(bkgdfj.m_jj, kde=False, norm_hist=True, bins=1000, color='blue', hist_kws={'weights': bkgdfj.weight, 'alpha': 0.7})
```

```python
plt.xlim(0,6)
sns.distplot(sigdfj.dEta_jj, kde=False, norm_hist=True, bins=100, color='green', hist_kws={'weights': sigdfj.weight, 'alpha': 0.7})
sns.distplot(bkgdfj.dEta_jj, kde=False, norm_hist=True, bins=100, color='blue', hist_kws={'weights': bkgdfj.weight, 'alpha': 0.7})
```

```python
plt.xlim(0,10)
sns.distplot(sigdf.N_valid_jets, kde=False, norm_hist=True, bins=10, color='green', hist_kws={'weights': sigdf.weight, 'alpha': 0.7})
sns.distplot(bkgdf.N_valid_jets, kde=False, norm_hist=True, bins=10, color='blue', hist_kws={'weights': bkgdf.weight, 'alpha': 0.7})
```

```python
plt.xlim(0,10)
sns.distplot(sigdf.N_valid_bjets, kde=False, norm_hist=True, bins=10, color='green', hist_kws={'weights': sigdf.weight, 'alpha': 0.7})
sns.distplot(bkgdf.N_valid_bjets, kde=False, norm_hist=True, bins=10, color='blue', hist_kws={'weights': bkgdf.weight, 'alpha': 0.7})
```

```python
plt.xlim(0,1000)
plt.ylim(0,4)
plt.xlabel('m_jj')
plt.ylabel('dEta_jj')
plt.title('signal')
plt.hexbin(sigdfj.m_jj, sigdfj.dEta_jj, gridsize=100, bins='log', C=sigdfj.weight.apply(lambda x: x*1e2), reduce_C_function=np.sum, cmap='inferno')
```

```python
plt.xlim(0,1000)
plt.ylim(0,4)
plt.title('background')
plt.xlabel('m_jj')
plt.ylabel('dEta_jj')
ax = plt.hexbin(bkgdfj.m_jj, bkgdfj.dEta_jj, gridsize=100, bins='log', C=bkgdfj.weight.apply(lambda x: x*1e-2), reduce_C_function=np.sum, cmap='inferno')
```
